```python
# Source - https://stackoverflow.com/questions/20906474/import-multiple-csv-files-into-pandas-and-concatenate-into-one-dataframe
# Posted by Gaurav Singh
# Retrieved 11/5/2025, License - CC-BY-SA 4.0

import pandas as pd
import os

all_files = os.listdir()

li = []

print(all_files)

for filename in all_files[:-2]:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
frame.drop_duplicates(subset="Tweet ID",inplace=True)
frame.describe()

# frame.to_csv("X_Tweets_Aggregated.csv")

```
